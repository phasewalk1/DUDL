```python
import numpy as np
import torch
```

```python
# transpose np arrays
nM = np.array([[1,2,3,4], [5,6,7,8]])
print(nM), print(' ')

# transpose it
print(nM.T)
```

```python
# using pytorch
tv = torch.tensor([ [ 1,2,3,4], [5,6,7,8] ])
print(tv), print(' ')

# transpose it
tvT = tv.T
print(tvT)
```

```python
# examine data types
print(f'Variable nM is of type {type(nM)}')
print(f'Variable tv is of type {type(tv)}')
```
