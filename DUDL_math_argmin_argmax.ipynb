```python
import numpy as np
import torch
import torch.nn as nn
```

```python
# create a vector
v = np.array([ 1,40,2,-3] )

# find and report the min/max values
minval = np.min(v)
maxval = np.max(v)
print('Min,max: %g,%g' %(minval,maxval))

# now for the argmin/max
minidx = np.argmin(v)
maxidx = np.argmax(v)
print('argmin,argmax: %g,%g' %(minidx,maxidx))
```

# With a Matrix

```python
M = np.array([ [0,1,10],
               [20,8,5]
               ])
print(M), print()

# various minima in this matrix!
minvals1 = np.min(M)
minvals2 = np.min(M,axis=0)
minvals3 = np.min(M,axis=1)
print(f'Min of entire matrix: {minvals1}')
print(f'Min of each column: {minvals2}')
print(f'Min of each row: {minvals3}')

# argmin/max
minidx1 = np.argmin(M)
minidx2 = np.argmin(M,axis=0)
minidx3 = np.argmin(M,axis=1)
```
